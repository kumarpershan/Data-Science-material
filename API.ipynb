```python
import requests
```

```python
url = "https://jsonplaceholder.typicode.com/users"
respose = requests.get(url)
print('Request Status', respose.status_code)
```

```python
data = respose.json()
```

```python
data[0:2]
```

```python
for user in data[0:2]:
    print('Name: ' , user['name'], '| Email: ', user['email'], "Phone Number " , user['phone'])
```

```python
import requests
city = "Karachi"
url = f"https://wttr.in/{city}?format=j1"
response = requests.get(url)
```

```python
print('Request Status', response.status_code)
```

```python
weather = response.json()
```

```python
weather
```

```python
temp =weather["current_condition"][0]["temp_C"]
desc = weather["current_condition"][0]["weatherDesc"][0]["value"]
print(f"Weather in {city}: {temp}°C, {desc}") 
```

```python
import requests
url = "https://www.thesportsdb.com/api/v1/json/3/searchteams.php"
params = {"t": "Arsenal"} # team name
response = requests.get(url, params=params)
data = response.json()
print(data)
```

```python
import base64
url = "https://postman-echo.com/basic-auth"
user, pwd = "postman", "password"
token = base64.b64encode(f"{user}:{pwd}".encode()).decode()
headers = {"Authorization": f"Basic {token}"}
response = requests.get(url, headers=headers)
print(response.status_code)
print(response.json())
```
