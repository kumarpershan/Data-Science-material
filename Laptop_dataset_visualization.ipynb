```python
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\Admin\\Downloads\\laptop_pricing_dataset_base.csv")
df.head()
```

```python
df.rename(columns = {'Acer': 'manufactur',
                    '4':'GPU', 
                    'IPS Panel':'screen',
                    '2':'GPU2', '1': 'OS',
                    '5':'CPU_core',
                    '35.56':'Screen_Size_cm',
                    '1.6':'CPU_frequency',
                    '8':'RAM_GB',
                    '256':'Storage_GB_SSD',
                    '1.6.1':'Weight_pounds',
                    '978':'Price'}, inplace=True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
x = df['manufactur'].unique()
y = df.groupby('manufactur')['GPU'].sum()
plt.bar(x, y)
plt.xlabel('Manufactur')
plt.ylabel('GPU count')
plt.xticks(rotation=90)
plt.show()
```

```python
x= df['manufactur'].unique()
y = df['Price'].sum()
plt.bar(x, y)
plt.xticks(rotation=90)
```
