```python
x = np.linspace(0, 10, 100)  # plot(x,y, '--')

plt.plot(x, np.sin(x), '-', label='Sine')
plt.plot(x, np.cos(x), '--', label='Cose');
fig = plt.figure()
```

# LINE PLOT

## Simple Line Plot

```python
%matplotlib inline
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
plt.grid(True)
```

```python
plt.plot(x, np.sin(x), color='red');
```

```python
plt.plot(x, x+2, color='cyan', linestyle='-', label='x+2')
plt.plot(x, x+5, color='green', linestyle='-.', label='x+5')
plt.plot(x, x+7, color='orange', linestyle='--', label='x+7')
plt.plot(x, x+9, color='red', linestyle=':', label='x+9')
plt.legend()
plt.savefig('Line Plots.png')
plt.show()
```

```python
plt.plot(x, x+2, '-c')
plt.plot(x, x+4, ':g')
plt.plot(x, x+6, '--o')
plt.plot(x, x+8, '-.v')
```

### Limit X and Y Values

```python
plt.plot(x, np.sin(x), '--c')
plt.xlim(2, 8)
plt.ylim(-1.5, 1.5);
```

### Title and Labels

The plt.axis() method allows you to set the x and y limits with a single call, by passing a list that specifies ============= [xmin, xmax, ymin,ymax]

```python
plt.plot(x, np.sin(x), )
plt.axis([-1, 11, -1.5, 1.5]);
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
plt.fill_between(x, np.sin(x), color='skyblue', alpha=0.4) 
```

### Short Way to write all Parameters 

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.legend(loc='upper left', fontsize=12, title='Legend')
plt.xlabel('x')
plt.ylabel('sin(x) and Cos(x)')
plt.title('Sin and Cos Curve')
```

The plt.axis() method goes even beyond this, allowing you to do things like auto
matically tighten the bounds around the current plot

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

 It allows even higher-level specifications, such as ensuring an equal aspect ratio so
 that on your screen, one unit in x is equal to one unit in y

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
plt.title('Example of an “equal” layout, with units matched to the output resolution')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randn(1,100)
plt.hist(data, bins = 10)
plt.grid(True)
```

The hist() function has many options to tune both the calculation and the display;

```python
plt.hist(data, bins=30, alpha=0.6,
         histtype='stepfilled', color='steelblue', edgecolor='black');
plt.grid(True)
plt.figure(figsize=(4,4))
```

```python
plt.style.use('classic')
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40)
plt.hist(x1,**kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```
