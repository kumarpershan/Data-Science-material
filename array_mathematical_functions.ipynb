# Array arithmetic

```python
import numpy as np
```

```python
x = np.arange(4)
print('x =', x)
```

```python
print('x +5 = ', x+5)
print('x-5 = ', x-5)
print('x*5 = ', x*5)
print('x/5 = ', x/5)
```

### OR Using Functions
All of these arithmetic operations are simply convenient wrappers around specific
functions built into NumPy; for example, the + operator is a wrapper for the add
function:

```python
print('x +5 = ', np.add(x,5))
print('x-5 = ', np.subtract(x,5))
print('x*5 = ', np.multiply(x, 5))
print('x/5 = ', np.divide(x, 5))
print('x**5 = ', np.power(x, 5))
print('x/5 = ', np.mod(x, 5))
```

## Absolute value
Just as NumPy understands Python’s built-in arithmetic operators, it also understands
Python’s built-in absolute value function:n:

```python
x = np.array([-2,-4, -1,0,1,-6])
print(x)
print('absolute x = ', abs(x))
```

The corresponding NumPy ufunc is np.absolute, which is also available under the
alias np.abs:

print('absolute x =', np.absolute(x))

 OR

```python
print('absolute x =', np.abs(x))
```

###### This ufunc can also handle complex data, in which the absolute value returns the 
magnitude:

```python
x = np.array([3 - 4j, 4 - 3j, 2 + 0j, 0 + 1j])
np.abs(x)
```

# Trigonometric functions

```python
theta = np.array([0, 30,45,60,90,180])
print("theta      = ", theta)
print("sin(theta) = ", np.sin(theta))
print("cos(theta) = ", np.cos(theta))
print("tan(theta) = ", np.tan(theta))
```

# Aggregations:

```python
a = np.arange(1,6)
a
```

```python
a1 = np.arange(11, 23).reshape(3,4)
a1
```

#### sum(array)----- sum of all elements in array

```python
print('the sum of array',a )
np.sum(a)
```

```python
print('the sum of array', np.sum(a1) )
```

### sum(array, axis)
#### axis=1 means row wise sum


```python
print('the sum of array', np.sum(a1, axis=1) )
```

#### axis=0 means column wise sum

```python
print('the sum of array', np.sum(a1, axis=0) )
```

### min(array)
====> return minimum value from array

```python
np.min(a1)
```

##### min(array, axis)
axis=0 means column wise
axis=1 means row wise

```python
print('row wise minimum', np.min(a1, axis=1))
```

```python
print('column wise minimum', np.min(a1, axis=0))
```

### max(array)
return maximum value from array

```python
np.max(a1)
```

## Similarly all aggregate functions work

```python
print('Min: ', np.min(a1))
print('Max: ', np.max(a1))
print('Mean: ', np.mean(a1))
print('Standard deviation: ', np.std(a1))
print('Variance: ', np.var(a1))
print('Median: ', np.median(a1))
print('Product: ', np.prod(a1))
print('Percentile: ', np.percentile(a1,2))
```

#### np.any(condition)
if condition satisfy for atleast 1 elemnet it return true

```python
np.any(a1>13) # we have atleast 1 element in array a1 whose vale is greater than 13
```

### np.all(condition)
if all elements satisfy condition it return True else False

```python
np.all(a1>13)# we have all element in array a1 whose vale is greater than 13
```

```python
np.all(a1>13)# we have all element in array a1 whose vale is greater than 13
```
