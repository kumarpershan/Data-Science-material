```python
def add(x,y):
    z = x + y
    return z

def subtract(x,y):
    z = x-y
    return z
def multiply(x,y):
    z = x*y
    return z
def divide(x,y):
    z = x/y
    return z
subtract(5,1)
```

```python
def calculator(x,y):
    print('Select any Operation +, -, *, /')
    ope = input('  ')
    if ope == '+':
        print(add(x,y))
    elif ope == '-':
        print(subtract(x,y))
    elif ope == '*' :
        print(multiply(x,y))
    elif ope == '/':
        print(divide(x,y))
    else:
        print('Enter a correct operator') 
    
calculator(10,2)
```

```python
calculator(10,2)
```
