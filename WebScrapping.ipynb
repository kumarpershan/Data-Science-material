```python
url = 'http://quotes.toscrape.com'
response = requests.get(url)
print('Request Status', response.status_code)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
heading = soup.find('h1')
print(heading.text)
```

```python
links = soup.find_all('a')
count = 0
for link in links:
    print(count, link.get('href'))
    count+=1
```

```python
quotes =soup.find_all('div',class_='quote') 
data = []
for i in quotes:
    text = i.find('span', class_='text').get_text(strip=True)
    author = i.find('small', class_='author').get_text(strip=True)
    data.append({
        "quote": text,
        "author": author})
print(data[:2]) 
```

```python
import csv
with open("DataQuote.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=["quote", "author"])
    writer.writeheader()
    writer.writerows(data)
```
