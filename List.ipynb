```python
l1 = [23,45,12,50,34,37,19,11]
l2=[]
l3 = []
l4= []
for i in l1:
    if i%2 == 0 and i%5 != 0:
        l2.append(i)
    elif i%5 == 0:
        l3.append(i)
    else:
        l4.append(i)
print(l2)
print(l3)
print(l4)
    
```

```python
l1 = [23,45,12,50,34,37,19,11]
l2=[]
l3 = []
l4= []
for i in l1:
    if i%5 == 0:
        l2.append(i)
    elif i%2 == 0:
        l3.append(i)
    else:
        l4.append(i)
print(l2)
print(l3)
print(l4)
```

#### List Comprehension

```python
s = []
for i in range(11, 16):
    s.append(i**2)
print(s)
```

```python
s1 = [i**2 for i in range(1,10)]
s1
```

```python
s2 = [x for x in range(1, 51) if x%5==0]
s2
```
