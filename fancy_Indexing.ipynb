## Fancy Indexing

```python
a  = np.arange(5, 15)
a
```

```python
[a[2], a[-1], a[0]]
```

```python
ind = [2,-1,0]
a[ind]
```

```python
a = np.arange(10, 35).reshape(5,5)
a
```

```python
# 25
# 10...
# 20...
[a[3], a[0], a[2]]
```

```python
row = [3, 0, 2]
a[row]
```

```python
a[  : , [3,0,4]]
```

```python
a
```

```python
28,22,26,32,13
row = [3, 2, 3,-1, 0 ]
col = [3, 2, 1, 2, 3]
a[row, col]
```

```python
row = [0,1,2,3,4]
col = [4,3,2,1,0]
a[row, col]
```

```python
a[[1,1,2,2], [1,2,1,2]]
```

```python
a[[1,1,2,2], [1,2,1,2]].reshape(2,2)
```

```python
a[1:3, 1:3]
```

```python
a[2,2] = 2
```

```python
a
```

```python
a = np.arange(11, 20)
a
```

```python
i = [1, 3, -3]
a[i] = 20
a
```

```python
i = [0,2,4]
a[i] = 10,30,50
a
```

```python
a = np.arange(10, 35).reshape(5,5)
a
```

```python
row = [1]
col = [2,3]
a[row, col] = 60, 70
a
```

```python
a[1, 2:4] = 60, 70 
```
