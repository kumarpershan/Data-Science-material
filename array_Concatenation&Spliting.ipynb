# Splitting of arrays
The opposite of concatenation is splitting, which is implemented by the functions
np.split, np.hsplit, and np.vsplit. For each of these, we can pass a list of indices
giving the split points:

```python
x = [1, 2, 3, 99, 99, 3, 2, 1, 6,7]
x1, x2, x3 = np.split(x, [3, 5])
print(x1, x2, x3)
```

```python
x = [2,4,5,6,8,9,11,4,5,10,12,15,16,18]
```

```python
x1, x2, x3, x4 = np.split(x, [3, 7, 9])
print(x1, x2, x3, x4)
```

### 2D array Split
split(array, indexes, axis=)
axis = 0 ====> vsplit
axis = 1 ====> hsplit

```python
a = np.arange(16).reshape((4, 4))
a
```

```python
x,y = np.split(a, 2)
print(x)
print('..............')
print(y)
```

```python
x,y = np.split(a, 2,axis=1)
print(x)
print('..............')
print(y)
```

###### np.vsplit used tp split array vertically 
syntax====> np.vsplit(array, [index1, index2, ...indexn])

```python
upper, lower = np.vsplit(a, [1])
print(upper)
print('..................')
print(lower)
```

```python
upper,middle, lower = np.vsplit(a, [1,3])
print(upper)
print('..................')

print(middle)
print('..................')
print(lower)
```

```python
x =  np.vsplit(a, [1])
x
```

```python
left, right = np.hsplit(a, [2])
print(left)
print('.........................')
print(right)
```

```python
left,middle, right = np.hsplit(a, [2,3])
print(left)
print('.........................')
print(middle)
print('.........................')
print(right)
```
