# 1. LINE PLOT (Matplotlib)

Use When: You want to show trends over time (e.g., stock prices, temperature over days).

Data Type: Continuous data (X-axis is usually time).

Example: Sales per month, website traffic over a week.

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

plt.figure(figsize=(5, 5)) 
plt.plot(x, y, marker='o', color='blue', linestyle='--')
plt.title("Line Plot Example" , fontsize=16)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xticks(rotation=45)            # Rotate labels
plt.tight_layout()
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.show()
```

# 3. HISTOGRAM (Matplotlib)

Use When===> You want to show the distribution of a single numerical variable.

Data Type===> Numerical.

Example===> Distribution of ages, exam scores, income.

```python
# Purpose: To show distribution of a variable
data = np.random.randn(1000)

plt.hist(data, bins=10, color='purple', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

# 4. PIE CHART (Matplotlib)

Use When: You want to show part-to-whole relationship.

Data Type: Categorical data (with proportions).

Example: Market share, browser usage, budget breakdown.

```python
# Purpose: To show proportion of parts to whole
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 20, 9]
colors = ['gold', 'lightblue', 'lightgreen', 'pink']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import seaborn as sns
data = {
    'Math':    [78, 85, 90, 95, 88], # 'Math': [78, 85, 90, 95, 88],
    'Science': [80, 74, 58, 73, 90],
    'English': [70, 75, 78, 82, 77]
}
df = pd.DataFrame(data)
# Calculate correlation matrix
corr_matrix = df.corr()
# Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='Blues', linewidths=0.5)
plt.title("Heatmap of Student Scores")
plt.show()
```
