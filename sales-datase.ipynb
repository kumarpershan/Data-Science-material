## Unique values

```python
print(df['product'].unique())
```

```python
print(df['category'].unique())
```

```python
df['category'] = df['category'].apply(lambda x : 'Bags' if x == 'Bgas' else
                                     'Clothing' if x == 'Clohting' else 
                                      'Shoes' if x== 'Shoeses' else x)
```

#### OR

```python
df['category'] = df['category'].replace({
    'Clohting':'Clothing',
    'Shoeses': 'Shoes',
    'Bgas' : 'Bags'
})
```

## Apply function

```python
df['price_with_tax'] = df['price'].apply(lambda x: x * 1.17)
df.head(3)
```

## Convert date into y-m-d  format

```python
df['date'] = pd.to_datetime(df['date'])
# print(df['date'].dt.year)
# print(df['date'].dt.month)
df.head(8)
```

```python
# Select a single column
print(df['product'])

# Select multiple columns
print(df[['product', 'price']])

# Select a specific row by index
print(df.loc[1])

# Select by position
print(df.iloc[0, 2])   # Row 0, column 2 (category)
```

```python
# Add new column
df['discount'] = df['price'] * 0.10
print(df)

# Modify existing column
df['revenue'] = df['price'] * df['quantity']
print(df)

# Drop column
df.drop('discount', axis=1, inplace=True)
print(df)

# Rename columns
df.rename(columns={'product': 'item_name'}, inplace=True)
print(df)
```

```python
# Filter rows where price > 500
print(df[df['price'] > 500])

# Multiple conditions
print(df[(df['price'] > 100) & (df['category'] == 'Electronics')])
```

```python
# Save to CSV
df.to_csv('sales_data.csv', index=False)

# Read again
# new_df = pd.read_csv('sales_data.csv')
# print(new_df.head())
```
