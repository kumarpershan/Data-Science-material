## Subarrays as no-copy views

```python
x = np.arange(1,21).reshape(4,5)
x
```

```python
z=x[4: :-3, 3:0:-2]
z
```

```python
z[1,1] = 200
z
```

```python
x
```

```python
z
```

```python
z
```

```python
x
```

```python
y = x[1:3, 2:4].copy()
y
```

```python
y[0,1]=90
y
```

## Concatenation of arrays

```python
x = np.arange(3)
y = np.arange(5,10)
print(x)
print(y)
```

```python
np.concatenate([x,y])
```

```python
x = np.arange(1,5).reshape(2,2)
x
```

```python
np.concatenate([x,x])
```

```python
y = np.arange(9).reshape(3,3)
y
```

```python
np.concatenate([x,y])
```

```python
np.concatenate([y,x])
```
