# Creating Arrays 

## 1- From Python List

```python
a1 = np.array([1,2,3,4,5])
a1
```

```python
type(a1)
```

```python
a2 =  np.array([1, 2, 3, 4], dtype='float32')
print(a2)
type(a2)
```

## 2- Zeros Array

```python
zero = np.zeros(5)
zero
```

```python
zero1 = np.zeros((3,4))
zero1
```

```python
zero1.ndim
```

## 3- Ones Array

```python
one = np.ones(6)
one
```

```python
one1 = np.ones((3,5), dtype=int)
one1
```

## 4- Array with Same Value

```python
a3 = np.full(5,6)
print(a3)
a3.ndim
```

```python
a4 = np.full((3,3), 100)
a4
```

## 5- Identity Array

```python
eye = np.eye(3)
eye
```

```python
ide = np.identity(3)
ide
```

## 6- Array from arange method

```python
a5 = np.arange(3, 31,3)
a5
```

```python
a5 = np.arange(3, 31,3).reshape(5,2)
a5
```

## 7- Array from linspace Method
Create an array of five values evenly spaced between two numbers

```python
a7 = np.linspace(0, 100, 5)
a7
```

```python
a7 = np.linspace(1, 200, 10).reshape(5,2)
a7
```

## 8- Array from random Method
Create a 3x3 array of uniformly distributed random values between 0 and 1

```python
a8 = np.random.random(4)
a8
```

```python
a8 = np.random.random((3,4))
a8
```

## 9- Array from normal Method
Create a 3x3 array of normally distributed random values with mean 0 and standard deviation 1

```python
a9 = np.random.normal(0,1,5)
a9
```

```python
a9 = np.random.normal(0,1,(3,3))
a9
```

```python
a9.std()
```

## 10- Array of Random Integer
 Create a 3x3 array of random integers in the interval [0, 10)

```python
a10 = np.random.randint(0,10,4)
a10
```

```python
a10 = np.random.randint(10,100,(3,3))
a10
```
