# Computation on Arrays: Broadcasting

```python
import numpy as np
```

```python
a = np.array([0,1,2])
b = np.array([5,5,5])
print(a)
print(b)
```

```python
a + b
```

```python
a + 5
```

```python
M = np.ones((3,3))
M
```

```python
M + a
```

```python
c = np.array([0,2,5]).reshape(3,1)
c
```

```python
M + c
```

```python
x = np.arange(1,4)
y = np.arange(4,7).reshape(3,1)
print(x)
print(y)
```

```python
x + y
```

```python
a = np.ones((3,3))
b = np.arange(3)
print(a)
print(b)
```

```python
a + b
```
