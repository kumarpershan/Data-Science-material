Generate a code which take a list of numbers and provide square of these 
numbers in a new list.
Numbers = [1,2,3,4]

```python
sqr = []
def square(L):
    for i in L:
        sqr.append(i**2)
    print(sqr)
numbers = [1,2,3,4]
square(numbers)
```

Write a code which take list of numbers as in input from user and check 
generate a list of numbers which is divisible by 2. Those number which are 
divisible by 2 will be saved in new list.

```python
eve = []
def Even(L):
    for i in L:
        if i%2 == 0:
            eve.append(i)
    print(eve)
numbers = [1,2,3,4]
Even(numbers)
```

Write a code which take two list of numbers (l1,l2) as in input and find the 
common elements and save it in new list.
l1 = [1,2,3,4]   l2 = [1,2,8,9]


```python
l1 = [2,3,4,5,6,7,8,9]
l2 = [1,2,4,8,11,23,9]
l3 = []
for i in l1:
    if i in l2:
        l3.append(i)
print(l3)
```

Write a code which provides a dictionary of cubic values. Key is number, 
value is the cube as follows:
:
{1:1, 2:8, 3:27, ....}

```python
cube = {}
for i in range(11, 16):
    cube[i]=i**3
print(cube)
```

```python
cube = {}
for i in range(1,6):
    cube.update({i:i**3})
print(cube)
```
