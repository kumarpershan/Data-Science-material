```python
data = {
    'Name': ['Ali', 'Sara', 'Sarim', 'Kareem'],
    'Age' : [23,22, 32,45],
     'Maths' : [89, 87,98,88],
    'Physics': [87, 56,77,89],
    'Chemistry': [85, 67,74,65]
                  }
student = pd.DataFrame(data)
student
```

```python
student[['Name', 'Chemistry']]
```

```python
student['Age'] = [23,22,21,20]
student
```

## Adding new column

```python
student['Python'] = [78,82,84,91]
student
```

```python
student['Total Marks'] = student[['Maths', 'Physics', 'Chemistry', 'Python']].sum(axis=1)
student
```

```python
student['Percentage'] = (student['Total Marks']/400) *100
```

```python
student['Percentage'] = student['Total Marks'].apply(lambda x: (x/400) *100 )

```

```python
student
```

```python
student['Grade'] = student['Percentage'].apply(lambda x: 
                                             'A1' if x>=85 else 
                                            'A' if x>=80 else
                                             'B'  if x>=70 else
                                              'c'  if x>=60 else 'D')
student
```

```python
student['Python'] = student['Python'] + 2 
student
```
