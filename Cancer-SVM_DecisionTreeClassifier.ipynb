```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
```

```python
df = pd.read_csv("C:\\Users\\Administrator\\Desktop\\Waseela\\Datasets\\Cancer.csv")
df.head(3)
```

```python
df.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
df.columns
```

```python
df['diagnosis'].unique()
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':0, 'B':1})
df.head(2)
```

```python
x = df.drop('diagnosis', axis=1)
corr = x.corr()
corr
```

```python
from sklearn.preprocessing import MinMaxScaler
data=['radius_mean', 'texture_mean','smoothness_mean','symmetry_mean', 
         'fractal_dimension_mean','radius_worst', 'texture_worst',
        'compactness_worst',  'symmetry_worst', 
        'fractal_dimension_worst']
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[data])
```

```python
x = scaled
y = df['diagnosis']
```

```python
x[:, :4]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_text, y_train, y_test = train_test_split(x, y, test_size=0.2,
                                                   random_state=42)
```

# Support Vector Machine

```python
svm = SVC(kernel='linear', C=1.0)  # You can also try 'rbf' or 'poly' kernels
svm.fit(x_train, y_train)
```

```python
y_pred = svm.predict(x_text)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
tree = DecisionTreeClassifier(random_state=42)
tree.fit(x_train, y_train)
y_pred_tree = tree.predict(x_text)
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_tree)) 
```
